```python
#this is going to be a SpaceX-based object oriented programming practice project. For the project, I will have an 
#inventory of several SpaceX hardware parts in a rocket. I will write a program that keeps track of how many flights 
#each part can last and then the program will tell the engineers when the part needs to be replaced. To make the simulation
#more realistic, I will add an element of randomness in the simulation such that on random flights BEFORE the expiration 
#flight of a certain piece of hardware, the system will tell the engineers to replace the part early as something
#unexpectedly stressed the piece of hardware more than usual. 
```

```python
import random
```

```python
class Hardware:
    
    #here I initialize a set of hardware that will need to be replaced after x amount of flights
    def __init__(self):
        self.combustion_chamber = 0
        self.fuel_pump = 0
        self.turbine = 0
        self.fuel_hose = 0
        self.grid_fin = 0
        self.cc_max_flights = 5
        self.fp_max_flights = 10
        self.tb_max_flights = 4
        self.fh_max_flights = 15
        self.gf_max_flights = 20
```

```python
class RandomProblem:
    
    #here I will make a method that picks a random number from 1 to 5. This will determine which part must be 
    #replaced before its scheduled replacement flight
    def __init__(self):
        return
    
    #this method generates random number from 1 to 10
    def random_number(self):
        return random.randint(1,10)
```

```python
class Rocket:
    
    #this method simply just makes the rocket that will fly and asks the engineer for the rocket's name. It also 
    #initializes an object from the Hardware class so that this rocket will have its needed hardware.
    def __init__(self):
        self.name = input("Name the rocket:")
        self.hardware = Hardware()
        
    #this method will keep track of how many times each part has flown
    def FlightsFlown(self, name, num_flights):
        ff = "The {} have/has flown {} time(s)."
        ff = ff.format(name, num_flights)
        print(ff)
    
    #this method will keep track of how many flights left a particular part has
    def FlightsLeft(self, name, num_flights):
        fl = "The {} have/has {} flights left. If zero, replace immediately."
        fl = fl.format(name, num_flights)
        print(fl)
    
    #this method tells the program how to run and keeps track of the specific parts on the rocket!
    def fly_rocket(self):
        while True:
            
            message1 = "Type 'launch' to launch {}. Typle 'abort' to abort the launch."
            message1 = message1.format(self.name)
            response = input(message1)
            if response == "abort":
                break
            
            message2 = """
Congrats with your succesful flight on {}. Here is the breakdown after today's launch on the status of {}:"""
            message2 = message2.format(self.name, self.name)
            print(message2)
            
            self.randnumber = RandomProblem()
            self.entropy = self.randnumber.random_number()
            self.hardware.combustion_chamber += 1
            self.hardware.fuel_pump += 1
            self.hardware.turbine += 1
            self.hardware.fuel_hose += 1
            self.hardware.grid_fin += 1
            
            if self.entropy == 1:
                print("COMBUSTION CHAMBER CRACKED ON LANDING. REPLACE IMMEDIATELY.")
                self.hardware.combustion_chamber = 0
            else:
                self.FlightsFlown("combustion chamber", self.hardware.combustion_chamber)
                self.FlightsLeft("combustion chamber", self.hardware.cc_max_flights - self.hardware.combustion_chamber)
                
            if self.entropy == 3:
                print("FUEL PUMP IS PREFORMING TOO SLOW. REPLACE IMMEDIATELY.")
                self.hardware.fuel_pump = 0
            else:
                self.FlightsFlown("fuel pump", self.hardware.fuel_pump)
                self.FlightsLeft("fuel pump", self.hardware.fp_max_flights - self.hardware.fuel_pump)
            
            if self.entropy == 5:
                print("TURBINE LOOKS WORN DOWN. REPLACE IMMEDIATELY.")
                self.hardware.turbine = 0
            else:
                self.FlightsFlown("turbine", self.hardware.turbine)
                self.FlightsLeft("turbine", self.hardware.tb_max_flights - self.hardware.turbine)
            
            if self.entropy == 7:
                print("FUEL HOSE SPRUNG A LEAK. REPLACE IMMEDIATELY.")
                self.hardware.fuel_hose = 0
            else:
                self.FlightsFlown("fuel hose", self.hardware.fuel_hose)
                self.FlightsLeft("fuel hose", self.hardware.fh_max_flights - self.hardware.fuel_hose)
                
            if self.entropy == 9:
                print("GRID FIN HIT A BIRD WHILE LANDING. REPLACE IMMEDIATELY.")
                self.hardware.grid_fin = 1
            else:
                self.FlightsFlown("grid fins", self.hardware.grid_fin)
                self.FlightsLeft("grid fins", self.hardware.gf_max_flights - self.hardware.grid_fin)
            
            if self.hardware.combustion_chamber == self.hardware.cc_max_flights:
                self.hardware.combustion_chamber = 0
                print("REPLACE COMBUSTION CHAMBER IMMEDIATELY")    
            elif self.hardware.fuel_pump == self.hardware.fp_max_flights:
                self.hardware.fuel_pump = 0
                print("REPLACE FUEL PUMP IMMEDIATELY")
            elif self.hardware.turbine == self.hardware.tb_max_flights:
                self.hardware.turbine = 0
                print("REPLACE TURBINE IMMEDIATELY")
            elif self.hardware.fuel_hose == self.hardware.fh_max_flights:
                self.hardware.fuel_hose = 0
                print("REPLACE FUEL HOSE IMMEDIATELY")
            elif self.hardware.grid_fin == self.hardware.gf_max_flights:
                self.hardware.grid_fin = 0
                print("REPLACE GRID FINS IMMEDIATELY")
```

```python
rocket = Rocket()
rocket.fly_rocket()
```
